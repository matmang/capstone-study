```python
# func class

class aa:
    def __init__(self, name):
        self.name = name
        print('초기화됨')
    def hello(self,x):
        print("안녕" + self.name + x)

a = aa("하")
a.hello('세요')
```

```python
# numpy

import numpy as np

x = np.array([1.0, 2.0, 3.0])

print(x)
print(x/2)
print(type(x))

y = np.array([2.0, 3.0, 4.0])

print(x + y)
print(x * y)
print(x / y)
```

```python
# 2차원 배열

a = np.array([[1,2],[3,4]])
b = np.array([[3,0],[0,6]])
c = np.array([10,20])

print(a.shape)
print(a.dtype)
print(a)

print(a+b)
print(a*b)

print(a*c)
print((a*c)[0])
print((a*c)[0][1])

x = a.flatten()
print(x)
print(x>2)
```

```python
# matplotlib

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)", linestyle="--")
plt.xlabel("x")
plt.xlabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
